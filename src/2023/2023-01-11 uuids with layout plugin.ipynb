```python
import mysql.connector

db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="secret",
    port="3306",
    database="serlo",
    charset="latin1"
)

def query(sql):
    c = db.cursor()
    c.execute(sql)

    return c.fetchall()
```

```python
data = query("""
    select
        entity.id, entity_revision_field.value
    from entity
    join entity_revision on entity.current_revision_id = entity_revision.id
    join entity_revision_field on entity_revision_field.entity_revision_id = entity_revision.id
    join uuid on uuid.id = entity.id
    where uuid.trashed = 0 and entity_revision_field.field = "content"
""")
```

```python
import csv

uuids_with_layout = [[x[0]] for x in data if '"layout"' in x[1]]

with open('/tmp/uuids_with_layout_plugin.csv', 'w') as f:
    write = csv.writer(f)
    write.writerows(uuids_with_layout)

result = "# Serlo-Inhalte mit dem Layout Plugin\n"

for x in uuids_with_layout:
    uuid = x[0]
    result += f"* [http://serlo.org/{uuid}](http://serlo.org/{uuid})\n"

from IPython.display import display, Markdown

display(Markdown(result))
```
