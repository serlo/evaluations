```python
import pandas as pd
import mysql.connector

from IPython.display import display, Markdown, HTML

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="secret",
    port="3306",
    database="serlo",
    charset="latin1"
)
```

```python
def fix_encoding(title):
    return bytes(title, "latin1").decode("utf8") if title != None else None

d = pd.read_sql("""
    select
        user.username as profile_username,
        user.id as profile_id,
        comment.date as comment_date,
        comment.content as comment_content,
        comment_user.username as comment_username,
        answer.date as answer_date,
        answer.content as answer_content,
        answer_user.username as answer_username
    from user
    join comment on comment.uuid_id = user.id
    left join comment as answer on answer.parent_id = comment.id
    join user as comment_user on comment_user.id = comment.author_id
    left join user as answer_user on answer_user.id = answer.author_id
    order by comment.id desc
""", db)

d["comment_content"] = d["comment_content"].apply(fix_encoding)
d["answer_content"] = d["answer_content"].apply(fix_encoding)
d["answer_delay"] = d["answer_date"] - d["comment_date"]
d["answer_delay_seconds"] = d["answer_delay"].dt.total_seconds()
d.to_csv("/tmp/threads_on_profiles.csv")
d
```
